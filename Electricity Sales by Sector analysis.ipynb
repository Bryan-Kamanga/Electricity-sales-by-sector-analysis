```python
#finding the residential sales by year
residential_sales = [sales.groupby('Year')['Residential  '].sum()]
print(residential_sales)                     
```

```python
#Commercial sales by year
Commercial_sales_by_year = (sales.groupby('Year')['Commercial  '].sum())
Commercial_sales_by_year
```

```python
#industrial sales by year
Industrial_sales_by_year = (sales.groupby("Year")["Industrial  "].sum())
Industrial_sales_by_year
```

```python
#transportation sales by year
Transportation_sales_by_year = (sales.groupby("Year")["Transportation  "].sum())
Transportation_sales_by_year
```

```python
#total sales by year
Total_sales_by_year = (sales.groupby("Year")["Total  "].sum())
Total_sales_by_year
```

```python
#checking the correlation analysis
sales[["Total  ",
       "Residential  ",
       "Commercial  ",
       "Industrial  ",
       "Transportation  ",]].corr()
```

```python
#checking the trend of Residential sales
plt.plot(sales['Year'], sales['Residential  '], color ="Orange")
plt.xlabel("Year")
plt.ylabel("Total Residential Sales")
plt.show()
```

```python
#checking the commercial sales trend
plt.plot(sales['Year'] , sales['Commercial  '], color = "blue")
plt.xlabel('Year')
plt.ylabel("Total Commercial Sales")
plt.show()
```

```python
#checking the trend of industrial sales
plt.plot(sales['Year'], sales['Industrial  '], color = "Red")
plt.xlabel("Year")
plt.ylabel("Total Industrial Sales")
plt.show()
```

```python
#Checking the Transportation sales trend
plt.plot(sales["Year"], sales["Transportation  "], color = "Green")
plt.xlabel("Year")
plt.ylabel("Total Transportation Sales")
plt.show()
```

```python
#growth rate by year
growth_rate = sales.set_index('Year')[['Residential  ','Commercial  ','Industrial  ','Transportation  ']].pct_change() * 100
growth_rate
```

```python
#checking the sector ranking, which sector contributes the most
sales.set_index('Year')[['Residential  ','Commercial  ','Industrial  ','Transportation  ']].rank(axis=1, ascending=False)
```
